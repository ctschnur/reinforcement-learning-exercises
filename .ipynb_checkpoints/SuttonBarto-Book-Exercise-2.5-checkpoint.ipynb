Design and conduct an experiment to demonstrate the difficulties that sample-average methods have for nonstationary problems. Use a modified version of the 10-armed testbed in which all the $q_*(a)$ start out equal and then take independent random walks (say by adding a normally distributed increment with mean zero and standard deviation 0.01 to all the $q_*(a)$ on each step). Prepare plots like Figure 2.2 for an action-value method using sample averages, incrementally computed, and another action-value method using a constant step-size parameter, $\alpha = 0.1$. Use $\epsilon = 0.1$ and longer runs, say of 10,000 steps.

```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import random

# shifted normal distributions

def gaussian_normal_dist(x, mu, sigma): 
    return 1/(sigma * np.sqrt(2 * np.pi)) * np.exp( - (x - mu)**2 / (2 * sigma**2) )

def bandit(a, mu, sigma): 
    """ sample reward distribution (here: normal distribution) of action a """
    # samples from a shifted normal distribution
    # mu = mus[a]
    # sigma = sigmas[a]
    sample_x = None
    while True: 
        half_range = 4.*sigma 
        sample_x = mu - half_range + 2. * half_range * np.random.rand()
        sample_y_max = gaussian_normal_dist(mu, mu, sigma)
        sample_y = np.random.rand() * sample_y_max
        if sample_y < gaussian_normal_dist(sample_x, mu, sigma): 
            return sample_x
        
def draw_true_values_and_estimates(a_arr, true_action_values, sigmas, Q):
    fig, axes = plt.subplots(len(a_arr), figsize=(6, 10), sharex=True, sharey=True)

    x = np.linspace(-5, 5)
    
    left, width = .25, .5
    bottom, height = .25, .5
    right = left + width
    top = bottom + height
    
    for i in a_arr: 
        axes[i].plot(x, gaussian_normal_dist(x, true_action_values[i], sigmas[i]))
        axes[i].text(right, top, "a = {:d}".format(i),
            horizontalalignment='right',
            verticalalignment='top',
            transform=axes[i].transAxes)
        
        # draw true values
        axes[i].axvline(true_action_values[i], label="true value")
        
        # draw estimates
        axes[i].axvline(Q[i], ls="--", label="estimate")
        # axes[i].legend()
        
    axes[0].legend()
    axes[-1].set_xlabel("Reward")
    axes[-1].set_xlim(-5, 5)
    
    fig.tight_layout()
    
# draw_true_values_and_estimates(a_arr, mus, sigmas, Q)

def run_bandit_problem(k, epsilon, time_steps): 
    """ """
    # k = 10  # k-armed bandit
    q_star_a = np.array([0.] * k)
    
    # mus and sigmas
    sigma_0 = 1.
    mu_0 = 0.
    sigmas = np.array([sigma_0] * k) 
    # mus = np.array([mu_0] * k)
    mus = np.random.normal(mu_0, sigma_0, k)
    
    N = np.array([0.] * k)
    Q = np.array([0.] * k)
    
    a_arr = np.array(range(k))  # a starts at 0, ends at k-1
    
    draw_true_values_and_estimates(a_arr, mus, sigmas, Q)
    
    # rf algorithm
    # time_steps = 1000
    # epsilon = 0.1
    
    rewards = []
    timesteps = np.array(range(time_steps))
    
    for t in timesteps: 
        a = None
        rnum = np.random.rand()
        
        # if rnum > epsilon:
        a = np.argmax(Q)  # returns the first! index of the argmax, if there are multiple equal ones
            # optionally, check for equal estimates and draw randomly among equal ones
        # else: 
        #     a = np.random.choice(a_arr)
            
        R = bandit(a, mus[a], sigmas[a])
        N[a] = N[a] + 1
        Q[a] = Q[a] + 1./N[a] * (R - Q[a])
      
        # collect reward
        rewards.append(R)
        
    draw_true_values_and_estimates(a_arr, mus, sigmas, Q)

    fig, ax = plt.subplots(1)
    
    ax.plot(timesteps, rewards)
    ax.set_xlabel("t")
    ax.set_ylabel("Rewards")
    
# k, epsilon, time_steps
run_bandit_problem(10, 0.1, 1000)
```

As we can see, at $t=0$, all $Q$ 's are 0, and they develop over time, such that at $t=1000$, one of the highest action-values is exploited ($a=4$), however, not the actual highest ($a=7$) in this particular case.

As we can see, in this particular non-stationary case with randomly walking reward distributions, the 'Percentage of optimal action value' is not steadily increasing, but starts to decrease fairly early on. The next version is to try an constant step-size factor instead, i.e. a 'exponential recency-weighted average'.   

## Non-stationary problem

Don't use sample averages, but constant step-size parameter $\alpha=0.1$, use non-greedy selection $\epsilon=0.1$ and more steps, i.e. 10.000 steps.   

```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import random
import itertools

%matplotlib notebook

# shifted normal distributions

def gaussian_normal_dist(x, mu, sigma): 
    return 1/(sigma * np.sqrt(2 * np.pi)) * np.exp( - (x - mu)**2 / (2 * sigma**2) )

def bandit(a, mu, sigma): 
    """ sample reward distribution (here: normal distribution) of action a """
    # samples from a shifted normal distribution
    # mu = mus[a]
    # sigma = sigmas[a]
    sample_x = None
    while True: 
        half_range = 4.*sigma 
        sample_x = mu - half_range + 2. * half_range * np.random.rand()
        sample_y_max = gaussian_normal_dist(mu, mu, sigma)
        sample_y = np.random.rand() * sample_y_max
        if sample_y < gaussian_normal_dist(sample_x, mu, sigma): 
            return sample_x
        
def draw_true_values_and_estimates(a_arr, true_action_values, sigmas, Q):
    fig, axes = plt.subplots(len(a_arr), figsize=(6, 10), sharex=True, sharey=True)

    x = np.linspace(-5, 5)
    
    left, width = .25, .5
    bottom, height = .25, .5
    right = left + width
    top = bottom + height
    
    for i in a_arr: 
        axes[i].plot(x, gaussian_normal_dist(x, true_action_values[i], sigmas[i]))
        axes[i].text(right, top, "a = {:d}".format(i),
            horizontalalignment='right',
            verticalalignment='top',
            transform=axes[i].transAxes)
        
        # draw true values
        axes[i].axvline(true_action_values[i], label="true value")
        
        # draw estimates
        axes[i].axvline(Q[i], ls="--", label="estimate")
        # axes[i].legend()
        
    axes[0].legend()
    axes[-1].set_xlabel("Reward")
    axes[-1].set_xlim(-5, 5)
    
    fig.tight_layout()
    
# draw_true_values_and_estimates(a_arr, mus, sigmas, Q)

def run_bandit_problem(k, epsilon, timesteps_arr, rewards_collector, count_of_optimal_action_collector): 
    """ """
    # k = 10  # k-armed bandit
    q_star_a = np.array([0.] * k)
    
    alpha = 0.1
    
    # mus and sigmas
    sigma_0 = 1.
    mu_0 = 0.
    sigmas = np.array([sigma_0] * k) 
    # mus = np.array([mu_0] * k)
    mus = np.random.normal(mu_0, sigma_0, k)
    
    N = np.array([0.] * k)
    Q = np.array([0.] * k)
    
    a_arr = np.array(range(k))  # a starts at 0, ends at k-1
    
    # draw_true_values_and_estimates(a_arr, mus, sigmas, Q)
    
    # rf algorithm
    # time_steps = 1000
    # epsilon = 0.1
    
    # rewards = []
    # timesteps_arr = np.array(range(time_steps))
    
    for t in timesteps_arr: 
        a = None
        rnum = np.random.rand()
        
        # change all 'true' action-values, i.e. the mus by a normally-distributed random walk
        sigma_1 = 0.01
        mus += np.random.normal(0., sigma_1, k)
        
        if rnum > epsilon:
            a = np.argmax(Q)  # returns the first! index of the argmax, if there are multiple equal ones
            # optionally, check for equal estimates and draw randomly among equal ones
        else: 
            a = np.random.choice(a_arr)
            
        R = bandit(a, mus[a], sigmas[a])
        N[a] = N[a] + 1
        Q[a] = Q[a] + alpha * (R - Q[a])
      
        # collect reward
        # rewards.append(R)
        rewards_collector[t] = R
        
        if a == np.argmax(mus): 
            count_of_optimal_action_collector[t] += 1
    
    # rewards_collector = np.array(rewards)
    # print("rewards_collector: ", np.shape(rewards_collector))
        
    # draw_true_values_and_estimates(a_arr, mus, sigmas, Q)

    # fig, ax = plt.subplots(1)
    # 
    # ax.plot(timesteps_arr, rewards)
    # ax.set_xlabel("t")
    # ax.set_ylabel("Rewards")
    
# k, epsilon, time_steps
k = 10
# epsilon = 0.1
epsilons = [#0., 
            #0.01, 
    0.1
]
timesteps_number = 10000
timesteps_arr = np.array(range(timesteps_number))
# rewards_collector
num_of_runs = 2000
# num_of_runs = 2000  # takes too long with unoptimized implementation

fig, [ax, ax2] = plt.subplots(2, figsize=(8, 6), sharex=True)
mpl_color_cycle = itertools.cycle(["green", "red", "blue"])

for epsilon in epsilons: 
    rewards_sum = np.zeros(timesteps_number)
    count_of_optimal_action_collector_sum = np.zeros(timesteps_number)
    for i in range(num_of_runs): 
        rewards_new = np.zeros(timesteps_number)
        count_of_optimal_action_collector_new = np.zeros(timesteps_number)
        run_bandit_problem(k, epsilon, timesteps_arr, rewards_new, count_of_optimal_action_collector_new)
        rewards_sum += rewards_new
        count_of_optimal_action_collector_sum += count_of_optimal_action_collector_new
        
        if i % int((float(num_of_runs)/float(10.))) == 0:
            print(i, "/", num_of_runs)
    
    rewards_average = rewards_sum / float(num_of_runs)
    count_of_optimal_action_collector_average = count_of_optimal_action_collector_sum / float(num_of_runs)

    color = next(mpl_color_cycle)
    ax.plot(timesteps_arr, rewards_average, color=color, label="epsilon={:.2f}".format(epsilon))
    ax2.plot(timesteps_arr, count_of_optimal_action_collector_average, color=color, label="epsilon={:.2f}".format(epsilon))

ax.set_xlabel("t")
ax.set_ylabel("Average of Rewards")
ax.legend()

ax2.set_xlabel("t")
ax2.set_ylabel("Percentage of optimal action")
ax2.set_ylim(0., 1.)
ax2.legend()

fig.tight_layout()
```

We can see, in the non-stationary problem, the percentage of optimal action performs better in the long run for the exponential recency-weighted average method (with constant time-step parameter) as compared to the sample-average method (with 1/n as time-step parameter). 

This is useful to keep in mind, since many RL problems are of a non-stationary nature and might perform better, if the most recent corrections to the estimates are weighted higher than older contributions.
